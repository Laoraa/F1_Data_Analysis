# Getting Started with FastF1 - Comparison of 2 laps by different drivers in a race

```python
import fastf1
fastf1.Cache.enable_cache('../data')
```

## 1. Session Selection

```python
session = fastf1.get_session(2025, 'Japan Grand Prix', 'R')
session.load()
```

## 2. Drivers Selection

```python
lap_leclerc = session.laps.pick_driver('LEC').pick_fastest()
lap_hamilton = session.laps.pick_driver('HAM').pick_fastest()

telemetry_leclerc = lap_leclerc.get_car_data().add_distance() 
telemetry_hamilton = lap_hamilton.get_car_data().add_distance()
```

## 3. Visualization

```python
from matplotlib import pyplot as plt
import fastf1.plotting

fastf1.plotting.setup_mpl()
```

```python
fig, ax = plt.subplots()
ax.plot(telemetry_leclerc['Distance'], telemetry_leclerc['Speed'], label='Leclerc')
ax.plot(telemetry_hamilton['Distance'], telemetry_hamilton['Speed'], label='Hamilton')
ax.set_xlabel('Distance (m)')
ax.set_ylabel('Vitesse (km/h)')
ax.set_title('Ferrari Speed Comparison - Fastest Lap (2025 Japanese Grand Prix)')
ax.legend()
plt.show()
```
